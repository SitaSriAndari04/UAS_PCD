```python
# Mengimpor pustaka-pustaka yang diperlukan
import numpy as np  # Untuk operasi numerik dan array
import pandas as pd  # Untuk manipulasi dan analisis data
import matplotlib.pyplot as plt  # Untuk membuat visualisasi
import cv2  # Pustaka OpenCV untuk tugas-tugas pengolahan citra

# Membaca sebuah gambar dari file menggunakan OpenCV
plat1 = cv2.imread('/content/1.jpg', 50)  # '50' sebagai parameter kedua menentukan mode skala abu-abu

# Menerapkan deteksi tepi Canny ke gambar
edges = cv2.Canny(plat1, 100, 200)

# Menampilkan gambar asli dan gambar dengan tepi yang terdeteksi secara berdampingan menggunakan matplotlib
plt.subplot(121),plt.imshow(plat1,cmap = 'gray')
plt.title('Gambar Asli'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Gambar Edge'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Membaca gambar dari file menggunakan OpenCV dalam mode skala abu-abu
plat2 = cv2.imread('/content/2.png', 50)

# Menerapkan deteksi tepi Canny pada gambar plat2 dengan batas ambang 100 dan 200
edges = cv2.Canny(plat2, 100, 200)

# Menampilkan gambar asli plat2 dalam subplot kiri
plt.subplot(121),plt.imshow(plat2,cmap = 'gray')
plt.title('Gambar Asli'), plt.xticks([]), plt.yticks([])

# Menampilkan gambar hasil deteksi tepi pada plat2 dalam subplot kanan
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Gambar Edge'), plt.xticks([]), plt.yticks([])

# Menampilkan kedua subplot secara bersamaan
plt.show()
```

```python
# Membaca gambar dari file menggunakan OpenCV
plat1 = cv2.imread('/content/1.jpg')

# Menerapkan deteksi tepi Canny pada gambar plat1 dengan batas ambang 50 hingga 255, menggunakan pendekatan non-L2gradient
edges = cv2.Canny(plat1,50,255,L2gradient=False)

# Menampilkan gambar hasil deteksi tepi dalam skala abu-abu
plt.imshow(edges,cmap='gray')

# Menampilkan gambar menggunakan plt.show()
plt.show()
```
